```python
table_rows = []
lis_1 = ['delta', 'delta21_14', 'delta7', 'meta', 'total']
lis_2 = ['confirmed', 'deceased', 'recovered', 'tested', 'vaccinated1', 'vaccinated2', 'population']
lis_2 = ['confirmed', 'deceased', 'recovered', 'tested', 'vaccinated1', 'vaccinated2']

for i in states:
    if i == "TT" : continue
    all_districts = list(df[colm[3]][i].keys())

    for j in all_districts:
        row_dict = {'states': i, "districts" : j}

        for x in lis_1:
            if df[colm[3]][i][j].get(x, 0) == 0 :
                for y in lis_2:
                    row_dict['{}_{}'.format(x, y)] = None
            else:
                for y in lis_2:
                    if df[colm[3]][i][j][x].get(y, 0) == 0 :
                        row_dict['{}_{}'.format(x, y)] = None
                    else:
                        row_dict['{}_{}'.format(x, y)] = df["districts"][i][j][x][y]

        table_rows.append(row_dict)
pd.DataFrame(table_rows).to_csv("data//data_min_districts.csv", index=False)
```
